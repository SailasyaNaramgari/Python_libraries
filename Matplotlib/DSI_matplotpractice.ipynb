***PIE CHART***

```python
data = [23,38,19,33,46]
plt.pie(data)

labels=['Python','C','C++','PHP','Java']
plt.pie(data,labels=labels)

plt.show()
```

```python
plt.figure(figsize=(7, 6))
plt.pie(data)
plt.title('programming language survey', fontsize=16)
plt.legend(labels)

plt.show()
```

```python
#explode

explode = [0.0, 0.0, 0.0,0.0, 0.09]
plt.pie(data, explode=explode, labels=labels)
plt.show()
```

```python
#Colors and shadows

colors = ['yellow', 'blue', 'crimson','lightblue','violet']
plt.pie(data,
        explode=explode,
        labels=labels,
        colors=colors,
        shadow=True)
plt.show()
```

```python
#Display the value

plt.pie(data,
        explode=explode,
        labels=labels,
        colors=colors,
        autopct='%.2f%%',    # here we also add the % sign
        shadow=True)
plt.show()
```

***DOUGHNUT CHART***

```python
plt.pie(data,
        labels=labels,
        colors=colors,
        startangle=90,
        wedgeprops={'width': 0.4})
plt.title("Doughnut chart")
plt.show()
```

***HORIZONTAL BAR CHART***

```python
korea_scores=(55,56,58)
canada_scores=(58,53,55)
china_scores=(43,50,50)
franch_scores=(45,55,99)

index=np.arange(3)
bar_width=.2

plt.figure(figsize=(8, 5))
k1=plt.bar(index,korea_scores,bar_width,alpha=.9,label="Korea" , color="lightblue")
c1=plt.bar(index+bar_width,canada_scores,bar_width,alpha=.9,label="Canada",color="yellow")
ch1=plt.bar(index+bar_width*2,china_scores,bar_width,alpha=.9,label="China",color="pink")
f1=plt.bar(index+bar_width*3,franch_scores,bar_width,alpha=.9,label="Franch",color="purple")

plt.xticks(index+.6/2,('Mathematics','Reading','Science'))
plt.ylabel('Mean score in PISA in 2012',fontsize=14)
plt.xlabel('Subjects',fontsize=14)
plt.title('Test scores ',fontsize=20)
plt.grid(True)
plt.legend()
```

```python
years = ['2016', '2017', '2018', '2019']
sugar = [43, 50, 53, 70]
wheat = [57, 50, 47, 30]

plt.figure(figsize=(6, 4))

plt.bar(years, sugar, label='sugar', color="violet")
plt.bar(years, wheat, bottom=sugar, label='Wheat',color="lightgreen")

plt.xlabel('Years', fontsize=14)
plt.ylabel('Preference', fontsize=14)
plt.title(' Sugar/Wheat export', fontsize=20)
plt.legend()
plt.show()
```

```python
years = ['2016', '2017', '2018', '2019']
rice =  np.array([57, 50, 47, 30])
wheat =  np.array([33, 40, 43, 60])
mustard =  np.array([20, 36, 23, 25])

plt.figure(figsize=(8, 5))

plt.bar(years, rice, label='Rice',color="orange")
plt.bar(years, wheat, bottom=rice, label='Wheat',color="yellow")
plt.bar(years, mustard, bottom=rice+wheat, label='Mustard seeds',color="lightblue")

plt.xlabel('Years', fontsize=14)
plt.ylabel('Preference (%)', fontsize=14)
plt.title('Rice,Wheat & Mustard seeds export', fontsize=20)
plt.legend()
plt.show()
```

```python
#Plotting multiple datasets

my_data = [163, 163, 164, 170, 170, 172, 173, 190,190,180]
sili_data = [161, 172, 174, 175, 181, 183, 186, 190,180,185]
bins = [160, 170, 180, 190]
names = ["Male", "Female"]

plt.hist([my_data, sili_data], bins=bins, label=names)
plt.title("Heights of male and female")
plt.ylabel("Number of people")
plt.xlabel("Height in cm")

plt.legend()
plt.show()
```

```python
plt.hist([my_data, sili_data], bins=bins, label=names, stacked=True, edgecolor='black')
plt.title("Heights of male and female")
plt.ylabel("Number of people")
plt.xlabel("Height in cm")

plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style

style.use('ggplot')

fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')

x = [1,2,3,4,5,6,7]
y = [5,6,7,8,2,5,6]
z = [1,2,6,3,2,7,3]

x2 = [-1,-2,-3,-4,-5,-6,-7]
y2 = [-5,-6,-7,-8,-2,-5,-6]
z2 = [1,2,6,3,2,7,3]

ax1.scatter(x, y, z, c='r', marker='o')
ax1.scatter(x2, y2, z2, c ='m', marker='*')

ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_zlabel('z axis')

plt.show()
```
