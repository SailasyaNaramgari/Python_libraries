```python
#Line plot - https://seaborn.pydata.org/generated/seaborn.lineplot.html
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#using matplotlib
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
plt.plot(var, var_1)
plt.show()
```

```python
#using sns
import pandas as pd


var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

x_1 = pd.DataFrame({"var" : var, "var_1":var_1})

sns.lineplot(x="var", y="var_1", data = x_1 )

plt.show()
```

```python
#without dataframe
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]


sns.lineplot(x=var, y=var_1 )

plt.show()
```

```python
#BOXPLOT - https://seaborn.pydata.org/generated/seaborn.boxplot.html
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
sns.boxplot(x='day' , y='total_bill' , data=var , hue="sex" , palette='BuPu' , order=["Fri","Sat","Thur","Sun"] , 
            showmeans=True , meanprops={"marker":"o" , "markeredgecolor":"blue"} , linewidth=3 , orient="v")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot( x='total_bill' ,  y='day' , data=var , hue="sex" , palette='BuPu' , order=["Fri","Sat","Thur","Sun"] , 
            showmeans=True , meanprops={"marker":"o" , "markeredgecolor":"blue"} , linewidth=3 , orient="h")
plt.show()
```
